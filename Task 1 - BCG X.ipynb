## Creating Column 'Revenue Growth'

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue ($)'].pct_change() * 100
```

## Fetching the dataset after adding column 

```python
df
```

## Filling '0' to null values in 'Revenue Growth'

```python
df['Revenue Growth (%)'] = df['Revenue Growth (%)'].fillna(0)
```

## Fetching dataset again

```python
df
```

## Creating Column 'Net Income Growth'

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income ($)'].pct_change() * 100
```

## Fetching the dataset after adding column 

```python
df
```

## Filling '0' to null values in 'Net Income Growth'

```python
df['Net Income Growth (%)'] = df['Net Income Growth (%)'].fillna(0)
```

## Fetching the dataset after adding column 

```python
df
```

## Creating Column 'Profit Margin ($)'

```python
df["Profit Margin (%)"] = (df["Net Income ($)"] / df["Total Revenue ($)"]) * 100
```

## Creating Column 'Debt Ration'

```python
df["Debt Ratio"] = df["Total Liabilities ($)"] / df["Total Assets ($)"]
```

## Creating some visuals to create some impactful insights.

## Plotting Total Revenue Trend (2022 - 2024)

```python
plt.figure(figsize=(10,5))
for company in df["Company"].unique():
    subset = df[df["Company"] == company]
    plt.plot(subset["Fiscal Year"], subset["Total Revenue ($)"], marker="o", label=company)
plt.title("Total Revenue Trend (2022-2024)")
plt.xlabel("Fiscal Year")
plt.ylabel("Revenue ($ in millions)")
plt.legend()
plt.show()
```

## Apple and Microsoft dominate in scale, while Tesla lags far behind in revenue despite being a fast-growing company.

## Plotting Net Income Trend

```python
plt.figure(figsize=(10,5))
for company in df["Company"].unique():
    subset = df[df["Company"] == company]
    plt.plot(subset["Fiscal Year"], subset["Net Income ($)"], marker="o", label=company)
plt.title("Net Income Trend (2022-2024)")
plt.xlabel("Fiscal Year")
plt.ylabel("Net Income ($ in millions)")
plt.legend()
plt.show()
```

## Apple remains the most profitable, Microsoft shows strong gains, while Tesla struggles with declining earnings.

## Plotting visuals for Assets vs Liabilities

```python
plt.figure(figsize = (10,5))
df_2024 = df[df["Fiscal Year"] == 2024]
x = df_2024["Company"]
plt.bar(x, df_2024["Total Assets ($)"], label="Assets")
plt.bar(x, df_2024["Total Liabilities ($)"], alpha=0.7, label="Liabilities")
plt.title("Assets vs Liabilities (2024)")
plt.ylabel("Value ($ in millions)")
plt.legend()
plt.show()
```

## Microsoft holds the strongest balance sheet, Apple is heavily leveraged, and Tesla remains much smaller in scale.

## Plotting Operating Cash Flow Trend (2022- 2024)

```python
plt.figure(figsize = (10,5))
for company in df["Company"].unique():
    subset = df[df["Company"] == company]
    plt.plot(subset["Fiscal Year"], subset["Cash Flow from Operating Activities ($)"], marker="o", label=company)
plt.title("Operating Cash Flow Trend (2022-2024)")
plt.xlabel("Fiscal Year")
plt.ylabel("Cash Flow ($ in millions)")
plt.legend()
plt.show()
```

## Apple and Microsoft generate massive cash flows, while Teslas operating cash generation remains limited.

## Plotting Piechart for Revenue Contribution by Company (2024)

```python
df_2024 = df[df["Fiscal Year"] == 2024]

plt.figure(figsize=(6,6))
plt.pie(df_2024["Total Revenue ($)"], 
        labels=df_2024["Company"], 
        autopct="%.1f%%", 
        startangle=90, 
        colors=sns.color_palette("pastel"))
plt.title("Revenue Contribution by Company (2024)")
plt.show()
```

## Apple dominates 2024 revenue share, Microsoft follows, while Tesla contributes only a small fraction.

## Plotting Donut Chart for Net Income Contribution by Company

```python
plt.figure(figsize=(6,6))
plt.pie(df_2024["Net Income ($)"], 
        labels=df_2024["Company"], 
        autopct="%.1f%%", 
        startangle=90, 
        colors=sns.color_palette("muted"),
        wedgeprops={'width':0.4})
plt.title("Net Income Contribution by Company (2024)")
plt.show()
```

## Microsoft and Apple dominate profitability, while Tesla contributes minimally despite strong revenue growth.

## Renaming Column 'Profit Margin (%) to 'Profit Margin'

```python
df.rename(columns={"Profit Margin (%)": "Profit Margin"}, inplace=True)
```

## Plotting violin plot for Profit Margin (%) Distribution (2022-2024)

```python
plt.figure(figsize=(7,5))
sns.violinplot(data=df, x="Company", y="Profit Margin", palette="Set2")
plt.title("Profit Margin (%) Distribution (20222024)")
plt.ylabel("Profit Margin (%)")
plt.show()
```

## Apple and Microsoft enjoy consistently high margins, whereas Tesla shows much thinner profitability spread.

## Plotting barcharts for Average Operating Cash Flow (2022 - 2024)

```python
avg_cf = df.groupby("Company")["Cash Flow from Operating Activities ($)"].mean().reset_index()

plt.figure(figsize=(7,5))
sns.barplot(data=avg_cf, x="Company", y="Cash Flow from Operating Activities ($)", palette="coolwarm")
plt.title("Average Operating Cash Flow (20222024)")
plt.ylabel("Cash Flow ($ in millions)")
plt.show()
```

## Apple consistently generates the strongest cash flows, reinforcing its financial strength.

## Plotting visual for analysing Debt Ratio Trend (2022 - 2024)

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="Fiscal Year", y="Debt Ratio", hue="Company", marker="o")
plt.fill_between(df[df["Company"]=="Apple"]["Fiscal Year"], 
                 df[df["Company"]=="Apple"]["Debt Ratio"], alpha=0.2)
plt.title("Debt Ratio Trend (20222024)")
plt.ylabel("Debt Ratio")
plt.xlabel("Fiscal Year")
plt.show()
```

## Apple carries higher leverage compared to Microsoft and Tesla, indicating greater financial risk.
